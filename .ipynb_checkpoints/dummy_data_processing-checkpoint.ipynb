```python
import json
import csv
from textblob import TextBlob
import re
from nltk.stem.wordnet import WordNetLemmatizer
import random
```

```python
start_year = 2009
next_year = 2019
```

```python
condensed_data = dict()
for year in range(start_year, next_year): 
    with open('Trump_Twitter_Data/condensed_{0}.json'.format(year)) as json_data:
        condensed_data[year] = json.load(json_data)
        
```

```python
time_label = ['dayofweek', 'month', 'day', 'time', 'timezone', 'year']
i = 0
for year in range(start_year, next_year):
    for index in range(len(condensed_data[year])):
        time_data = condensed_data[year][index]['created_at'].split(' ')
        for index_time in range(len(time_label)):
            condensed_data[year][index][time_label[index_time]] = time_data[index_time]
        condensed_data[year][index].pop('created_at', None)
        #All timezone is equal to '+0000'
        condensed_data[year][index].pop('timezone', None)
```

```python
def get_topic(text):
    accept_tag = ['NN', 'NNS', 'NNP', 'NNPS']
    topic = []
    #result_original = []
    b = TextBlob(text)
    wordnet_lemmatizer = WordNetLemmatizer()
    for tag in b.tags:
        if(tag[1] in accept_tag):
            #result_original.append(tag[0])
            topic.append(wordnet_lemmatizer.lemmatize(tag[0]))
    return topic
```

```python
with open('information/countries.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        countries = row
for index, country in enumerate(countries):
    countries[index] = country.split('|')
countries_name = []
for country in countries:
    countries_name.append(country[1])
```

```python
with open('information/states_usa.txt') as file:
    content = file.read().splitlines()
states = []
num_of_states = int(content[0])
for i in range(num_of_states):
    states.append((content[1+i+num_of_states], content[1+i]))
states_name = []
for state in states:
    states_name.append(state[1])
```

```python
for year in range(start_year, next_year):
    for index in range(len(condensed_data[year])):
        text = condensed_data[year][index]['text']
        condensed_data[year][index]['retweet_of'] = re.findall('(?<=RT\s)\100.*?(?=:)', text)
        condensed_data[year][index]['mention'] = re.findall('@\w+(?=\s|$|:)', text)
        condensed_data[year][index]['links'] = re.findall('http.*?(?=\s|$)', text)
        condensed_data[year][index]['hashtags'] = re.findall('#\w+(?=\s|$|:|#)', text)
        condensed_data[year][index]['text'] = re.sub('http.*?(?=\s|$)', '', condensed_data[year][index]['text'])
        condensed_data[year][index]['text'] = re.sub('RT\s\100.*?:', '', condensed_data[year][index]['text'])
        if 'id_str' in condensed_data[year][index]: del condensed_data[year][index]['id_str']
        text = condensed_data[year][index]['text']
        condensed_data[year][index]['topics'] = get_topic(text)
```

```python
def get_random_emotion():
    emotion_value = random.uniform(0,1)
    if emotion_value < 0.4:
        return 'ANGER'
    elif emotion_value < 0.5:
        return 'FEAR'
    elif emotion_value < 0.7:
        return 'JOY'
    elif emotion_value < 0.75:
        return 'LOVE'
    elif emotion_value < 0.85:
        return 'SADNESS'
    else:
        return 'SURPRISE'
```

```python
for year in range(start_year, next_year):
    for index in range(len(condensed_data[year])):
        condensed_data[year][index]['states'] = []
        condensed_data[year][index]['countries'] = []
        topics = condensed_data[year][index]['topics']
        for topic in topics:
            if topic in states_name:
                if topic not in condensed_data[year][index]['states']:
                    condensed_data[year][index]['states'].append(topic)
            if topic in countries_name:
                if topic not in condensed_data[year][index]['countries']:
                    condensed_data[year][index]['countries'].append(topic)
        condensed_data[year][index]['emotion'] = get_random_emotion()
```
